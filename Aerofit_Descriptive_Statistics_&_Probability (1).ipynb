**DETECTING OUTLIERS**

To detect the outliers we can use boxplots

The quantitative variables in the dataset are :

1. Age   

2. Education  

3. Usage     

4. Fitness    

5. Income   

6. Miles

```python
fig, axis = plt.subplots(nrows=3, ncols=2,figsize=(6,5))
fig.subplots_adjust(top=1.7)
sns.boxplot(data=df,x='Age',ax=axis[0,0])
sns.boxplot(data=df,x='Education',ax=axis[0,1])
sns.boxplot(data=df,x='Usage',ax=axis[1,0])
sns.boxplot(data=df,x='Fitness',ax=axis[1,1])
sns.boxplot(data=df,x='Income',ax=axis[2,0])
sns.boxplot(data=df,x='Miles',ax=axis[2,1])
plt.show()
```

#Insights
* Age, Education and Usage have few outliers while Income and Miles have more outliers

To remove/clip the data between the 5 percentile and 95 percentile

```python
percentile_5 = np.percentile(df['Age'],5)
percentile_95 = np.percentile(df['Age'],95)
df_age = np.clip(df['Age'],percentile_5,percentile_95)
df_age
```

```python
percentile_5 = np.percentile(df['Education'],5)
percentile_95 = np.percentile(df['Education'],95)
df_edu = np.clip(df['Education'],percentile_5,percentile_95)
df_edu
```

```python
percentile_5 = np.percentile(df['Usage'],5)
percentile_95 = np.percentile(df['Usage'],95)
df_usage = np.clip(df['Usage'],percentile_5,percentile_95)
df_usage
```

```python
percentile_5 = np.percentile(df['Fitness'],5)
percentile_95 = np.percentile(df['Fitness'],95)
df_fit= np.clip(df['Fitness'],percentile_5,percentile_95)
df_fit
```

```python
percentile_5 = np.percentile(df['Income'],5)
percentile_95 = np.percentile(df['Income'],95)
df_income=np.clip(df['Income'],percentile_5,percentile_95)
df_income
```

```python
percentile_5 = np.percentile(df['Miles'],5)
percentile_95 = np.percentile(df['Miles'],95)
df_miles=np.clip(df['Miles'],percentile_5,percentile_95)
df_miles
```

```python
fig, axis = plt.subplots(nrows=3, ncols=2,figsize=(10,8))
fig.subplots_adjust(top=1.0,right=1)
sns.boxplot(data=df_age,orient='h',ax=axis[0,0])
sns.boxplot(data=df_edu,orient='h',ax=axis[0,1])
sns.boxplot(data=df_usage,orient='h',ax=axis[1,0])
sns.boxplot(data=df_fit,orient='h',ax=axis[1,1])
sns.boxplot(data=df_income,orient='h',ax=axis[2,0])
sns.boxplot(data=df_miles,orient='h',ax=axis[2,1])
plt.show()
```

Creating new columns for better understandings

Age - Groups

1. 18 - 29  : Young

2. 30 - 39 : Middle Age

3. 40 : 50 : Old


Education Column

1. Primary Education: upto 12

2. Secondary Education: 13 to 15

3. Higher Education: 16 and above


Income groups

1. 29000 - 50000 : Low

2. 51000 - 75000 : Medium

3. 76000 - 105000 : High


Miles column

1. Light Activity - Upto 50 miles

2. Moderate Activity - 51 to 100 miles

3. Active Lifestyle - 101 to 200 miles

4. Highly active - Above 200 miles

```python
#new columns using cut
df['Age group'] = pd.cut(df['Age'],bins=[17,29,39,50],labels=['Young','Middle Aged','Old'])
df['Education group'] = pd.cut(df['Education'],bins=[0,12,15,21],labels=['Primary','Secondary','Higher'])
df['Income group'] = pd.cut(df['Income'],bins = [29000,50000,75000,105000],labels=['Low','Medium','High'])
df['Miles group'] = pd.cut(df['Miles'],bins=[0,50,100,200,360],labels=['Light','Moderate','Active','Highly active'])
df.head()
```
